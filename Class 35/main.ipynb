# Function

```python
# 5! = 5*4*3*2*1 = 120
def facto(x):
    if x==0:
        return 1
    return x* facto(x-1)
```

```python
facto(5)
```

# Lambda Anonymous Function

lambda arguments: expression

```python
def sums(x,y):
    return x+y
sums(4,5)
```

```python
f = lambda x,y: x+y
f(4,5)
```

```python
f = lambda x,y: x*y
f(4,5)
```

```python
f = lambda x: x+x
f(4)
```

# Lambda in higher order

```python
def operation_higher(x,y, operation):
    return operation(x,y)

operation_higher(4,5, lambda x,y: x+y)
```

```python
def operation_higher(x,y, operation):
    return operation(x,y)
operation_higher(4,5, lambda x,y: x*y)
```
