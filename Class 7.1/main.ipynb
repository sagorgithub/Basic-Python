# If Else Statements in Python

```python
amount = 0

net_units = float(input('Enter your unit value : '))

if net_units <= 100:
    amonut = 0
    
    
elif net_units > 100 and net_units <= 200:
    amount = (net_units - 100) * 5
    

elif net_units>200:
    amount = ((net_units - 200) *10) + 500

print('amount is ',amount)
    
```
